# Printing Data

Printing or data output in Python refers to the process of displaying information or data on the screen or to a file. To print data in Python, we use the built-in "print" function. The print function allows you to display data of various types, such as strings, integers, and floating-point numbers.

In addition to the basic print function, Python also offers print formatting, which allows you to create more efficient and readable print statements. Print formatting allows you to add variables, control the width and precision of numbers, and align text.

For example, the f-string format allows you to embed variables directly into a string, making it easier to read and modify. The .format() method also allows you to insert variables into a string, with more control over the formatting.

````python
#Example 1: Using f-strings to embed variables in a string
name = "John"
age = 25
print(f"My name is {name} and I'm {age} years old.")

#Example 2: Using the .format() method to insert variables in a string
name = "Mary"
age = 30
print("My name is {} and I'm {} years old.".format(name, age))


#Example 3: Using string concatenation 
print("My name is " + name + " and I'm " + str(age) + " years old.")


#Example 4: Controlling the precision of floating-point numbers using format()
import math
pi = math.pi
print("The value of pi is approximately {:0>4.3f}".format(pi))

````

##Formatting printed variables

In Python, we can format printed variables to display their values in a specific way using a variety of formatting options. This allows us to control the appearance and precision of our printed output.

One way to format printed variables is by using the string formatting method, which involves using curly braces {} as placeholders for variables and the format() method to insert the values. For example, we can format a float variable to display only two decimal places.

Another way to format printed variables is by using f-strings, which is a newer and more concise method introduced in Python 3.6. This method involves prefixing a string with the letter "f" and using curly braces {} as placeholders for variables.

````python
#The following Python program demonstrates the use of the string modulo operator (%) to print fancy output.
#First, we print an integer using %d as a placeholder and passing in the integer value 102.
print("An integer number : %d" % (102))

#Then, we print a floating-point number using %.2f as a placeholder and passing in the floating-point value 5.333.
#The .2f specifies that we want to display two digits after the decimal point.
print("A float number : %.2f" % (5.333))

#Next, we print an octal number using the %o placeholder and passing in the integer value 25.
#The .3o specifies that we want to display the octal number with a precision of three digits.
print("%.3o" % (25))

#We also print the binary representation of the integer 25 using the built-in bin() function.
print(bin(25))

#Finally, we print a number in exponential notation using the %.3E placeholder and passing in the floating-point value 356.08977.
#The .3E specifies that we want to display the number in scientific notation with a precision of three digits.
print("%.3E" % (356.08977))

#Another example of using the string modulo operator to format strings:
name = "John"
age = 25
height = 1.75

#We can use %s as a placeholder for a string, %d for an integer, and %.2f for a floating-point number.
#Here, we concatenate multiple strings and format the output using the string modulo operator.
print("My name is %s. I am %d years old and %.2f meters tall." % (name, age, height))

#We can also use f-strings to format strings in Python 3.6 and later:
name = "Jane"
age = 30
weight = 62.5789

#We can use curly braces {} as placeholders for variables and prefix the string with f to create an f-string.
#Here, we format the output using an f-string.
print(f"My name is {name}. I am {age} years old and weigh {weight:.1f} kilograms.")

#The :.1f specifies that we want to display the weight with one digit after the decimal point.

````

#Type conversions

## Lists
This is the most common data structure in Python. A list is a collection of items that are ordered and mutable. Lists are created using square brackets and commas to separate the items.

One of the main advantages of lists in Python is that they can hold different types of data, including strings, integers, booleans, and even other lists. This flexibility makes them a versatile tool for managing and manipulating data in Python programs.

We can access the items in a list using the indexing operator (square brackets), where the first item has an index of 0. We can also use slicing to extract a portion of a list, which returns a new list with the selected items.

Using the indexing operator (square brackets) on the left side of an assignment, we can update one of the list items. This is because lists are mutable, meaning that we can change their contents after they have been created.


````python
# create a list with three elements (an integer, a string, and a boolean), and print the list
mylist = [4, 'Hi', True]
print(mylist)

# create an empty list, and print the list
mylist = []
print(mylist)

# create a list with multiple elements (integers, strings, and booleans)
mylist = [4, 'Hi', True, False, 84.3, 'Many', True]

# print the element at index 2 of the list
print(mylist[2])

# print the first three elements of the list
print(mylist[:3])

# print all elements of the list starting from index 2
print(mylist[2:])

# print the last two elements of the list
print(mylist[-2:])

# print the length of the list
print(len(mylist))
````

##Tuples
In Python, a tuple is a sequence of values that can contain items of any data type. They are very similar to lists, in that they allow items to be accessed by their index. However, tuples are different from lists because tuples are immutable objects, whereas lists are mutable.

This means that once a tuple is created, it cannot be changed. In contrast, a list can be modified by adding, removing, or changing items in the list. Tuples are more memory efficient than lists, since they require less memory to store the same amount of data.

When it comes to time efficiency, tuples have a slight advantage over lists, especially when it comes to lookups. This is because tuples are optimized for read-only operations, whereas lists are optimized for modifications.

Tuples are a good choice for storing data that is not meant to be changed, such as configuration settings, database connection strings, or file paths. However, if you need to modify the data frequently, a list would be a better choice.

##Dictionaries
A dictionary is a collection of key-value pairs that allows you to store and access data in an organized and efficient way. Dictionaries are ordered, meaning the items are arranged in a specific order, and changeable, which means you can modify, add, or delete elements. Additionally, dictionaries do not allow duplicate keys, but they do allow duplicate values.

To access a value in a dictionary, you can use its corresponding key. This makes it easy to quickly retrieve information, even from large sets of data. ## If … elif … else Conditionals

In Python, conditional statements allow the program to make decisions based on certain conditions. The if statement is used to execute a block of code if a certain condition is true. If the condition is false, the program will skip that block of code and move on to the next statement.

The elif statement can be used after the if statement to add additional conditions to check. If the initial if statement is false, the program will move on to check the elif statement. If the condition in the elif statement is true, the corresponding block of code will be executed.

The else statement is used to execute a block of code if none of the conditions in the previous if and elif statements are true. It acts as a catch-all statement.

Conditional statements in Python use indentation to define blocks of code, so it is important to use consistent and appropriate indentation.

````python
mynumbers = [3,8,7,14]

# Sort the list of numbers in ascending order
mynumbers.sort()

# Print the sorted list
print(mynumbers)

# Define a variable with a value of 9
a_number=5

# Check if a_number is in mynumbers
if a_number in mynumbers:
    # If it is, print a message indicating it is in the list
    print(a_number, " is in ", mynumbers)

# If a_number is not in mynumbers, check if it is larger than the largest element in the list
elif a_number > mynumbers[-1]:
    # If it is, print a message indicating it is not in the list and it is larger than the largest element
    print(a_number, "is not in ", mynumbers, " and it is larger than the largest element of ", mynumbers)

# If a_number is not larger than the largest element in the list, check if it is smaller than the smallest element in the list
elif a_number < mynumbers[0]:
    # If it is, print a message indicating it is not in the list and it is smaller than the smallest element
    print(a_number, "is not in ", mynumbers, " and it is smaller than the smallest element of ", mynumbers)

# If a_number is neither larger than the largest element nor smaller than the smallest element in the list
else:
    # Print a message indicating it is not in the list but it is within the range of the minimum and maximum elements
    print(a_number, "is not in ", mynumbers, " but it is within the range of the minimum and maximum elements of ", mynumbers)

# Short if statement: if a_number is in mynumbers, print a message indicating it is in the list
if a_number in mynumbers: print(a_number, " is in ", mynumbers)

# Short if else statement: if a_number is in mynumbers, print a message indicating it is in the list; otherwise, print a message indicating it is not in the list
print(a_number, " is in ", mynumbers) if a_number in mynumbers else print(a_number, " is not in ", mynumbers)

````

## While Loops

A while loop is a control flow statement in Python that allows for the repeated execution of a block of statements while a specified condition is true. The loop continues until the condition is no longer true. The syntax for a while loop is straightforward:

```
while condition:
  code to be executed
```

The loop will continue to execute as long as the condition is True. The condition can be any Boolean expression. Within the while loop block, you can include any number of statements to be executed repeatedly.

The break statement is used to stop the execution of the loop prematurely, regardless of whether the condition is still True. The continue statement skips the current iteration and continues to the next iteration of the loop.

Care should be taken when using while loops to avoid infinite loops, which can cause a program to crash. It is important to ensure that the loop condition will eventually become false to avoid this issue.

````python
# Define a list of numbers
mynumbers = [1, 2, 3, 8, 4, 7, 14]

# Sort the list in ascending order
mynumbers.sort()

# Print the sorted list
print(mynumbers)

# Initialize a counter variable k
k = 0

# Execute a while loop while k is less than the length of mynumbers
while k < len(mynumbers):
    # Print the k-th element of mynumbers
    print(mynumbers[k])
    # Increment k by 1
    k += 1

# Print a new line character
print('listo, terminó el bucle\n')
````

````python
# Reset the counter variable k to 0
k = 0

# Execute a while loop while k is less than the length of mynumbers
while k < len(mynumbers):
    # Check if the k-th element of mynumbers is equal to 8
    if mynumbers[k] == 8:
        # Increment k by 1 and skip to the next iteration of the loop
        k += 1
        continue
    # Print the k-th element of mynumbers
    print(mynumbers[k])
    # Increment k by 1
    k += 1

# Print a new line character
print('listo \n')
````

## Functions with variable arguments
These functions can accept an uncertain number of arguments. This type of function is useful when we are not sure about the number of arguments we need to pass to a function. To implement this in Python, we use the “wildcard” or “*” notation like this – "*args" OR "**kwargs" as our function’s argument.

The special syntax "*args" is used to pass a variable number of arguments to a function. It allows us to pass a non-keyworded, variable-length argument list. When the function is called, all the passed arguments are collected into a tuple named "args." We can then iterate over this tuple to access each argument. This makes the function flexible enough to handle any number of arguments.

The special syntax "**kwargs" is used to pass a keyworded, variable-length argument list. We use the name kwargs with the double star, and the double star allows us to pass through keyword arguments (and any number of them). When the function is called, all the keyword arguments are collected into a dictionary named "kwargs." We can then access each argument by its keyword.


````python
# Define a function called myFun that takes in a variable number of arguments
def myFun(*args):
    # Initialize a counter variable and an empty list
    i = 0
    printthis = []
    
    # Iterate over all the arguments passed in and append them to the list
    for arg in args:
        printthis.append(arg)
        
    # Print the list of arguments
    print(printthis)

# Call the myFun function with 2 string arguments
myFun('Hello', 'it\'s me ')
````

````python
# Define a function called myFun that takes in one argument and a variable number of arguments
def myFun(arg1, *args):
    # Print the first argument passed in
    print("First argument:", arg1)
    
    # Iterate over the variable number of arguments passed in and print them out
    for arg in args:
        print("Next argument through *args:", arg)

myFun('just one argument')
# Call the myFun function with 4 string arguments
myFun('Hello', 'Welcome', 'to', 'GeeksforGeeks')
````

````python
# Define a function called myFun that takes in a variable number of arguments and a variable number of keyword arguments
def myFun(*args, **kwargs):
    # Print out the variable arguments
    print("args:", args)
    
    # Print out the keyword arguments
    print("kwargs:", kwargs)
    print(kwargs.keys())
    print(kwargs.values())

# Call the myFun function with 3 string arguments as variable arguments and 3 keyword arguments
myFun('geeks', 'for', 'geeks', first="Geeks", mid="for", last="Geeks")
````

````python
# Define a function called myFun that takes in 3 arguments
def myFun(arg1, arg2, arg3):
    # Print out each argument
    print("arg1:", arg1)
    print("arg2:", arg2)
    print("arg3:", arg3)

# Create a tuple called args with 3 string elements
args = ("Geeks", "for", "Geeks")

# Call the myFun function with the args tuple unpacked using the * operator
myFun(*args)

# Create a dictionary called kwargs with 3 key-value pairs
kwargs = {"arg1": "Geeks", "arg2": "for", "arg3": "Geeks"}

# Call the myFun function with the kwargs dictionary unpacked using the ** operator
myFun(**kwargs)
````

#List Comprehension 
List comprehension in Python is a concise and powerful syntax to create a new list from an existing list or sequence, by applying an operation or a condition. The syntax of list comprehension is enclosed in square brackets, where the operation or expression that generates the new list is defined, followed by a for loop that iterates over the items in the sequence.

List comprehension is very useful in cases where we need to filter or transform a list, without having to write lengthy for loops and conditional statements. For example, we can create a list of even numbers from an existing list of numbers, or we can apply a mathematical operation to each element of a list.

One of the advantages of list comprehension is that it allows us to create lists in a more readable and concise way, making the code easier to understand and maintain. It is also a faster and more efficient way of generating new lists, compared to traditional for loops.

````python
# Using a for loop to create a list of squares
result = []
for i in range(1, 11):
    result.append(i ** 2)
print(result)

# Using list comprehension to create a list of squares
result = [i ** 2 for i in range(1, 11)]
print(result)
````

The use of list compression makes the code more compact and, above all, the execution much faster.

This structure can include conditional expressions that must be fulfilled so that an element can be included in the resulting list. The scheme of list compression with conditional expressions would be the following:

[operation for item in sequence if condition]

or 

[operation if condition else anotheroperation for x in sequence]

````python
result = [i**2 for i in range(1,11) if i % 3 == 0] 
print(result)

result = [i**2 if i % 3 == 0 else i for i in range(1,11) ] 
print(result)
````

For dictionaries, we need to define the key and value of each element that we insert into them. For this reason, the compression expression for dictionaries must include both concepts, the rest of the expression is the same as the previous one:

{key: value for item in sequence if condition}

 In the case of sets, we'll use the set creation symbols, followed by the same compression structure we used with lists:

{item for item in secuencia if condicion} 



````python
result = {i: i**2 for i in range(1,11) if i % 2 == 0} 
print(result)


result = {i**2 for i in range (1,11) if i % 2 == 0} 
print(result)
 
````

#Errors and Exceptions

There are situations where the interpreter cannot execute code as expected. Exceptions occur when errors occur at runtime. When an exception occurs, the program execution is stopped immediately, and the interpreter tries to find an exception handler that can handle the exception. The exception handler code is called an exception handler block. It is defined using the "try-except" block in Python.

A "try" block is used to enclose the code that could potentially throw an exception. The "except" block specifies what should happen when an exception occurs in the "try" block. It allows the program to catch the exception and continue execution. We can catch different types of exceptions by using multiple "except" blocks or a single block with multiple exceptions separated by commas.

In addition to "try-except," Python provides several other statements to handle exceptions, such as "finally," which defines a block of code that must be executed after the try block, regardless of whether an exception is thrown or not. Understanding how to handle errors and exceptions in Python is crucial to creating robust and reliable programs.

````python
# We define a list called "lista" with three integer elements
lista = [2, 3, 4]

# We execute a for loop from 0 to 6 (7 times)
for i in range(7):
    try:
        # We print the element of "lista" in position i
        print(lista[i])
    except:
        # If there is an error, we print a message indicating that the position does not exist
        print("No existe esa posición")
````

````python
# We define two lists: "lista1" with numbers from 0 to 19 and "lista2" with integers, a string, and an empty string
lista1 = [*range(20)]
lista2 = [2, 3, 4, "A", 27, ""]

# We execute a for loop for each element in "lista2"
for i in lista2:
    try:
        # We print the element of "lista1" in position i
        print(lista1[i])
    except IndexError:
        # If the index i is out of range, we print a message indicating that the position does not exist
        print("No existe esa posición")
    except TypeError:
        # If the index i is of a wrong type, we print a message indicating that the operation type is incorrect
        print("El tipo de la operación es erróneo")
    except:
        # If there is any other error, we print a message indicating that something went wrong
        print("Ha ocurrido otro error")
    else:
        # If there were no errors, we print a message indicating that everything went well
        print("todo bien")
    finally:
        # In any case, we print a message indicating that the try-except block has ended
        print("Hemos terminado el bloque try-except")
````

#OOP and Classes

There are several programming paradigms (ways of structuring the code, variables and functions in order to perform given activities). The two most common paradigms are Procedural Programming and Object-Oriented Programming (OOP). 

Object-oriented programming (OOP) is a programming paradigm that revolves around the concept of objects, which are instances of classes. In Python, classes are used to create objects with similar properties and behaviors. A class can be thought of as a blueprint for creating objects, defining the attributes and methods that they will have.

Attributes are variables that store data specific to each instance of the class, while methods are functions that define the behaviors of the objects. When a class is instantiated, a new object is created with the defined attributes and methods.

For example, if we want to create a class for a person, we can define attributes such as name, age, and gender, and methods such as walking or talking. These attributes and methods can then be accessed and modified by the object instance.

In Python, classes are defined using the class keyword followed by the class name, and the methods and attributes are defined within the class using the def keyword and self parameter. self refers to the instance of the class and is used to access its attributes and methods.

Creating an instance of a class is called instantiation, and it is done using the class name followed by parentheses. This creates a new object with the properties and methods defined in the class.

One of the key features of OOP is inheritance, which allows a new class to be based on an existing class. This means that the new class inherits all the attributes and methods of the existing class, and can also add its own attributes and methods.

Another key feature of OOP is encapsulation, which refers to the idea of hiding the implementation details of a class from the outside world. This means that the outside world only interacts with the class through its public interface, which is made up of its methods.

## sys module

*   sys.path: a list of directories Python searches for modules when importing.
*   sys.stdin, sys.stdout, and sys.stderr: standard input, output, and error streams.
*   sys.exit(): terminates the Python interpreter with an optional exit code.
*   sys.version: a string containing the Python version number and build information.



````python
import sys
"""
In this example, we use sys.version_info to check the Python version being used 
to run the script. If the version is less than 3.0, we print an error message 
and exit the script. This is useful for ensuring that the script is compatible 
with the required Python version.
"""
if sys.version_info < (3, 0):
    print("This script requires Python 3 or later.")
    sys.exit()

"""
In this example, we use sys.path to add a directory containing a custom module 
to the import path. We then import the module and use its functions in the
rest of the script.
"""
# Add a directory to the import path
sys.path.append('/path/to/my/module')
print(sys.path)

"""
The sys.platform function returns a string representing the platform on which
Python is running. The value of this string is determined at compile time and
is platform-specific. It is often used in scripts to conditionally execute
code based on the platform.
"""
print(sys.platform)  # print the name of the platform on which Python is running
# e.g., "win32" for Windows, "linux" for Linux, "darwin" for macOS, etc.

"""
The sys.getdefaultencoding() function returns the default string encoding used
by Python. This encoding is used to convert Unicode strings to bytes when
reading from or writing to files or communicating with other systems. The 
default encoding can be changed using the sys.setdefaultencoding()
function, although this is not recommended as it can cause compatibility
issues.
"""
print(sys.getdefaultencoding())  # print the default string encoding used by Python
# e.g., "utf-8", "ascii", etc.
````

##random module

The random module in Python provides a set of functions to generate random numbers and perform random selections. These functions are useful for a wide range of applications, including simulations, cryptography, games, and machine learning.

The module offers various types of distributions, such as uniform, normal, and exponential, to generate random numbers with different characteristics. It also includes functions to shuffle sequences, select random items, and generate random strings.

The random module uses a pseudorandom number generator that produces numbers based on a seed value. The seed value can be set manually or generated from the system clock, and it ensures that the same sequence of random numbers is produced every time the same seed is used. This feature makes the random numbers predictable, which is useful for testing and debugging purposes.

However, it is important to note that the random numbers generated by the random module are not truly random, but rather pseudorandom. This means that they are determined by a mathematical formula and can be predicted if the seed value is known. Therefore, the random module should not be used for cryptographic applications that require true randomness.

````python
import random

import random

# Generate a random integer between 0 and 10 (inclusive) and print it
print(random.randint(0, 10))

# Shuffle the elements in the list "my_list" and print the shuffled list
my_list = [1, 2, 3, 4, 5]
random.shuffle(my_list)
print(my_list)

# Choose a random element from the list "my_list" and print it
my_list = ["apple", "banana", "orange"]
print(random.choice(my_list))

# Generate a random float between 0 and 1 and print it
print(random.random())

# Choose a random character from the string "my_string" and print it
my_string = "hello world"
print(random.choice(my_string))

import string

# Generate a random password with 8 characters (letters and digits)
password = ''.join(random.choices(string.ascii_letters + string.digits, k=8))
print(password)

# Generate a list of 10 random integers between 0 and 100 (inclusive) and print it
my_list = [random.randint(0, 100) for _ in range(10)]

import datetime

# Generate a random date between January 1, 2022 and December 31, 2022 (inclusive) and print it
start_date = datetime.date(2022, 1, 1)
end_date = datetime.date(2022, 12, 31)
random_date = start_date + (end_date - start_date) * random.random()
print(random_date)

# Generate a random color and print it (as a tuple of RGB values)
color = (random.randint(0, 255), random.randint(0, 255), random.randint(0, 255))
print(color)

# Choose 3 random elements from the list "my_list" and print them
my_list = ["apple", "banana", "orange", "pear", "kiwi"]
sample = random.sample(my_list, 3)
print(sample)

# Generate a random boolean value (True or False) and print it
print(bool(random.getrandbits(1)))

# Choose a random element from a combination of three lists and print it
list1 = ["apple", "banana", "orange"]
list2 = ["dog", "cat", "bird"]
list3 = ["car", "bike", "bus"]
item = random.choice(list1 + list2 + list3)
print(item)
````

##Matrix multiplication
Multiplication of two matrices is defined if and only if the number of columns of the left matrix is the same as the number of rows of the right matrix. If A is an m-by-n matrix and B is an n-by-p matrix, then their matrix product AB is the m-by-p matrix whose entries are given by dot product of the corresponding row of A and the corresponding column of B.

\begin{align}
{\displaystyle [\mathbf {AB} ]_{i,j}=a_{i,1}b_{1,j}+a_{i,2}b_{2,j}+\cdots +a_{i,n}b_{n,j}=\sum _{r=1}^{n}a_{i,r}b_{r,j},}
\end{align}

###Series
Series are one-dimensional data structures, like lists or arrays in numpy. What differentiates the series is that the index of each element can be a label that we assign, similar to what is done in dictionaries. To create a series we use the Series() method where we will add the values and, if we want, the custom indices. We can access series by index, range or position. Series are based on numpy arrays. For this reason, we can easily apply vectorized operations, just like we did with arrays in numpy. However, it is important that both strings have the same indices, otherwise they may return unexpected results.

````python
import numpy as np
import pandas as pd

# Creating a pandas Series with custom index
my_series=pd.Series([1,2,3,4,5], index=['a', 'b', 'c', 'd', 'e'])
print(my_series)

# Creating a pandas Series from numpy array with custom index
series1=pd.Series(np.array([1, 2, 3, 4, 5]), index=['a', 'b', 'c', 'd', 'e'])
print(series1)

# Creating a pandas Series from a dictionary
a_dictionary={'a':7,'b':8,'c':9}
series2=pd.Series(a_dictionary)
print(series2)
````

````python
# Accessing elements of a Series
print(series1['a'])  # Accessing element with index 'a'
print(series1[3])   # Accessing element at position 3
print(series1[:2])  # Accessing elements up to index 2 (excluding)

# Accessing index and values of a Series
print(series1.index)   # Displaying the index of the Series
print(series1.values)  # Displaying the values of the Series

# Displaying the Series
print(series1)
print(series2)
````

````python
# Arithmetic operations between two Series
result=series1+series2
print(result)

# Mathematical operations on a Series
print(series1*5)        # Multiplying each element by 5
print(np.sqrt(series1)) # Taking the square root of each element in the Series
````

A histogram takes many data points and separates them into bins, or ranges of values. The heights of the bars represent the number of data points that fall into that bin. Histograms are a good way to visually summarize the data.

**Density plot:** 

A density plot is a smoothed, continuous version of a histogram estimated from the data. The most common form of estimation is known as kernel density estimation. In this method, a continuous curve (the kernel) is drawn at every individual data point and all of these curves are then added together to make a single smooth density estimation. The kernel most often used is a Gaussian (which produces a Gaussian bell curve at each data point).

# Interpolating data

````python
from scipy.interpolate import interp1d
import numpy as np

xs = np.arange(0,10,1)
print(xs)
ys = 2*np.sin(xs)
print(ys)

plt.figure()
plt.plot(xs,ys,'o')

newxs=np.arange(0, 9, 0.1)
interp_func = interp1d(xs, ys, kind="linear")
newys = interp_func(newxs)
plt.plot(newxs,newys,'-r')

interp_func = interp1d(xs, ys, kind="quadratic")
newys = interp_func(newxs)
plt.plot(newxs,newys,'--g')

interp_func = interp1d(xs, ys, kind="nearest")
newys = interp_func(newxs)
plt.plot(newxs,newys,'.k')


````
